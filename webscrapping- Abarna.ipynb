# importing libraries

```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.thoughtco.com/largest-cities-in-india-1435045"
```

# website response

```python
res=requests.get(url)
```

```python
res
```

# parsing

```python
soup=BeautifulSoup(res.text)
```

```python
soup
```

```python
city_names=soup.select("td:nth-child(2)")
```

```python
city_names
```

```python
 metropoliton_population=soup.select("td:nth-child(4)")
```

```python
metropoliton_population
```

```python
cityproper_population=soup.select("td:nth-child(5)")
```

```python
cityproper_population
```

```python
territory=soup.select("td:nth-child(3)")
```

```python
territory
```
